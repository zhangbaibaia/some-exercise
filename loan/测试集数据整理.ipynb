# bank数据特征构建并填充缺失值

```python
df_bank_test = pd.read_csv('df_bank_test.csv')
```

```python
df_user_test=pd.read_csv('df_user_test.csv')
```

```python
df_bank_test.shape
```

```python
df_user_test.shape
```

```python
df = pd.merge(left = df_bank_test, right= df_user_test, how = 'outer',on = 'user_id')
df.shape
```

```python
len(df.user_id.unique())    #共有13899个用户
```

```python
df.head()
```

```python
t=df.drop(['time','sex','job','education','marriage','residence'],axis=1)
t.head()
```

```python
gb1=t[(t['loan_type']==0)].groupby(["user_id"],as_index=False)#收入统计
gb2=t[(t['loan_type']==1)].groupby(["user_id"],as_index=False)#支出统计
gb3=t[(t['income']==1)].groupby(["user_id"],as_index=False)#工资收入统计
x1=gb1['loan_amount'].agg({'earningnum' : 'count','earningsum':'sum'})  #收入笔数   收入总额
x2=gb2['loan_amount'].agg({'expendnum' : 'count','expendsum':'sum'})    #支出笔数    支出总额
x3=gb3['loan_amount'].agg({'incomenum' : 'count','incomesum':'sum'})    #工资收入笔数    工资收入总额
```

```python
df_bank_test1=pd.merge(t, x1,how='left', on = "user_id")
df_bank_test2=pd.merge(df_bank_test1, x2,how='left', on = "user_id")
df_bank_test3=pd.merge(df_bank_test2, x3,how='left', on = "user_id")
df_bank_test3.head()
```

```python
df_bank_test=df_bank_test3.drop(["Unnamed: 0"],axis=1)
```

```python
df_bank_test['earningnum-expendnum']=df_bank_test3['earningnum']-df_bank_test3['expendnum']
df_bank_test['earningsum-expendsum']=df_bank_test['earningsum']-df_bank_test['expendsum']
df_bank_test['un_incomenum']=df_bank_test['earningnum']-df_bank_test['incomenum']
df_bank_test[' un_incomesum']=df_bank_test['earningsum']-df_bank_test['incomesum']
df_bank_test['incomenum*(earningnum-expendnum )']=df_bank_test['incomenum']*df_bank_test['earningnum-expendnum']
df_bank_test[' incomesum*(earningsum-expendsum )']=df_bank_test['incomesum']*df_bank_test['earningsum-expendsum']
```

```python
df_bank_test.head()
```

```python
df_bank_test=df_bank_test.drop(['loan_type','loan_amount','income'],axis=1)
df_bank_test.shape
```

```python
df_bank_test=df_bank_test.drop_duplicates()
df_bank_test.shape
```

```python
df_bank_test.isnull().sum()
```

```python
df_bank_test.mean()
```

```python
bank_test=df_bank_test.fillna(df_bank_test.mean())
```
